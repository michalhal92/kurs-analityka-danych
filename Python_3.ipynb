Obsługa błędów

```python
print("Próba podziału przez zero")
result = 10 / 0
print(result)
```

```python
try:
    print("Próba podziału przez zero")
    result = 10 / 0
    print(result)
except ZeroDivisionError:
    print("Nie można dzielić przez zero")
else:
    print("Dzielenie zakończone sukcesem,result is", result)
finally:
    print("Ten kod wykonywany jest zawsze")
```

```python
# Obsługa wielu wyjątków:

try:
    value = int(input("Podaj dzielnik dla 10: "))
    result = 10 / value
    print(result)
except ValueError:
    print("Podano nieprawidłowy typ danych.")
except ZeroDivisionError:
    print("Próba dzielenia przez zero.")
```
