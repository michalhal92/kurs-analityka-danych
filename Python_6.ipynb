```python
# Po≈ÇƒÖczenia danych

df1 = pd.DataFrame({
    'Key1': ['K0', 'K1', 'K2', 'K3'],
    'Key2': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'Key1': ['K0', 'K1', 'K2', 'K3'],
    'Key2': ['K0', 'K1', 'K2', 'K3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
print(df1)
print()
print(df2)
```

```python
# ≈ÅƒÖczenie DataFrame'√≥w na podstawie kolumny 'Key1'
result = pd.merge(df1, df2, on='Key1')
print(result)
```

```python
# ≈ÅƒÖczenie po wielu kolumnach
result_multiple_columns = pd.merge(df1, df2, on=['Key1', 'Key2'], how = 'inner')
result_multiple_columns
```

```python
# R√≥znica miedzy join i merge
# ≈ÅƒÖczenie DataFrame'√≥w przy u≈ºyciu indeksu
result = df1.join(df2, lsuffix='_left', rsuffix='_right')
print(result)
```

```python
### ZAAWANSOWANE

# Po≈ÇƒÖczenia danych

df1 = pd.DataFrame({
    'Key1': ['K0', 'K1', 'K2', 'K3', 'K4'],
    'Key2': ['K0', 'K1', 'K2', 'K3', 'K4'],
    'A': ['A0', 'A1', 'A2', 'A3', 'A4'],
    'B': ['B0', 'B1', 'B2', 'B3', 'A4']
})

df2 = pd.DataFrame({
    'Key1': ['K0', 'K1', 'K2', 'K3', 'K5'],
    'Key2': ['K0', 'K1', 'K2', 'K3', 'K5'],
    'C': ['C0', 'C1', 'C2', 'C3', 'C5'],
    'D': ['D0', 'D1', 'D2', 'D3', 'D5']
})
```

```python
print(df1)
print()
print(df2)
```

```python
#Dlaczego merge jest  bardziej intuicyjny ni≈º join ;-)


#df.join():
#Domy≈õlnie ≈ÇƒÖczy po indeksie, a nie po kolumnach.
#Domy≈õlny how='left' ‚Äì to prawda, ale tylko przy ≈ÇƒÖczeniu po indeksach.
#Je≈õli chcesz po≈ÇƒÖczyƒá po kolumnach, musisz u≈ºyƒá parametru on=....

#≈ÅƒÖczenie nastƒôpuje po indeksach
df1.join(df2, lsuffix='_left', rsuffix='_right')
#print(result) << printowany by≈Ç result stƒÖd nieporozumienie, dlaczego wyglƒÖda≈Ço jakby by≈Ç join :)
```

```python
 pd.merge(df1, df2, on=['Key1', 'Key2'], how = 'left')
```

```python
#Nastƒôpne zajƒôcia
#1. jak wyswietlic okienko z definicja funkcji/metody

#Wymu≈õ tooltip dokumentacji:
#Umie≈õƒá kursor na nazwie funkcji (np. merge) i naci≈õnij:

#Ctrl + Shift + Space (Windows)
#Cmd + Shift + Space (Mac)


#2. dlaczego join zachowuje sie jednak jak inner a nie left
#df.join() ≈ÇƒÖczy po indeksie, wiƒôc Pandas spr√≥buje dopasowaƒá indeksy df1 i df2, co w Twoim przypadku nie ma sensu, bo Key1 i Key2 to kolumny, a nie indeksy.
#≈ªeby join zadzia≈Çal tak samo jak merge ( z how='left') nale≈ºy jeszcze podzia≈Çaƒá na indeksie:
df1.join(df2.set_index(['Key1', 'Key2']), on=['Key1', 'Key2'], how='left')
#Na zajƒôciach nie zauwa≈ºy≈Çam, ≈ºe po join printowa≈Ça mi siƒô zmienna 'result' a nie wynik joina, stƒÖd nieporozumienie dot. inner/left joina.
```
