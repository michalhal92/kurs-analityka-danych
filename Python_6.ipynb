```python
# Połączenia danych

df1 = pd.DataFrame({
    'Key1': ['K0', 'K1', 'K2', 'K3'],
    'Key2': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'Key1': ['K0', 'K1', 'K2', 'K3'],
    'Key2': ['K0', 'K1', 'K2', 'K3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
print(df1)
print()
print(df2)
```

```python
# Łączenie DataFrame'ów na podstawie kolumny 'Key1'
result = pd.merge(df1, df2, on='Key1')
print(result)
```

```python
# Łączenie po wielu kolumnach
result_multiple_columns = pd.merge(df1, df2, on=['Key1', 'Key2'], how = 'inner')
result_multiple_columns
```

```python
# Róznica miedzy join i merge
# Łączenie DataFrame'ów przy użyciu indeksu
result = df1.join(df2, lsuffix='_left', rsuffix='_right')
print(result)
```

```python
### ZAAWANSOWANE

# Połączenia danych

df1 = pd.DataFrame({
    'Key1': ['K0', 'K1', 'K2', 'K3', 'K4'],
    'Key2': ['K0', 'K1', 'K2', 'K3', 'K4'],
    'A': ['A0', 'A1', 'A2', 'A3', 'A4'],
    'B': ['B0', 'B1', 'B2', 'B3', 'A4']
})

df2 = pd.DataFrame({
    'Key1': ['K0', 'K1', 'K2', 'K3', 'K5'],
    'Key2': ['K0', 'K1', 'K2', 'K3', 'K5'],
    'C': ['C0', 'C1', 'C2', 'C3', 'C5'],
    'D': ['D0', 'D1', 'D2', 'D3', 'D5']
})
```

```python
print(df1)
print()
print(df2)
```

```python
#Dlaczego merge jest  bardziej intuicyjny niż join ;-)


#df.join():
#Domyślnie łączy po indeksie, a nie po kolumnach.
#Domyślny how='left' – to prawda, ale tylko przy łączeniu po indeksach.
#Jeśli chcesz połączyć po kolumnach, musisz użyć parametru on=....

#Łączenie następuje po indeksach
df1.join(df2, lsuffix='_left', rsuffix='_right')
#print(result) << printowany był result stąd nieporozumienie, dlaczego wyglądało jakby był join :)
```

```python
 pd.merge(df1, df2, on=['Key1', 'Key2'], how = 'left')
```

```python
#Następne zajęcia
#1. jak wyswietlic okienko z definicja funkcji/metody

#Wymuś tooltip dokumentacji:
#Umieść kursor na nazwie funkcji (np. merge) i naciśnij:

#Ctrl + Shift + Space (Windows)
#Cmd + Shift + Space (Mac)


#2. dlaczego join zachowuje sie jednak jak inner a nie left
#df.join() łączy po indeksie, więc Pandas spróbuje dopasować indeksy df1 i df2, co w Twoim przypadku nie ma sensu, bo Key1 i Key2 to kolumny, a nie indeksy.
#Żeby join zadziałal tak samo jak merge ( z how='left') należy jeszcze podziałać na indeksie:
df1.join(df2.set_index(['Key1', 'Key2']), on=['Key1', 'Key2'], how='left')
#Na zajęciach nie zauważyłam, że po join printowała mi się zmienna 'result' a nie wynik joina, stąd nieporozumienie dot. inner/left joina.
```
